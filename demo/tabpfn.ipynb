# Load Model

```python
from dcat_ap_hub import Dataset

url = "http://localhost:8081/datasets/97834246-63d9-4ad8-bc2d-9c4f2d022049.jsonld"

ds = Dataset.from_url(url)
raws = ds.download(data_dir="./data", force=True)
```

```python
print(ds)
list(raws)
```

```python
model = ds.load_model()
```

```python
from sklearn.metrics import accuracy_score


model.fit(X_train[:2], y_train[:2])

# Predict labels
predictions = model.predict(X_test)


print("Accuracy", accuracy_score(y_test, predictions))
```
