# Load Dataset

```python
from dcat_ap_hub import Dataset

url = "https://ki-daten.hlrs.de/de/dataset/https-piveau-io-set-data-predictive-maintenance-ttl"

ds = Dataset.from_url(url)
ds.download(data_dir="./data")
files = ds.process(processed_dir="./processed")
```

```python
X_train = files["X_train.npy"].data
y_train = files["y_train.npy"].data
X_test = files["X_test.npy"].data
y_test = files["y_test.npy"].data

print(f"Train shape: X={X_train.shape}, y={y_train.shape}")
print(f"Test shape: X={X_test.shape}, y={y_test.shape}")
```

# Load Model

```python
from dcat_ap_hub import Dataset

url = "https://ki-daten.hlrs.de/hub/repo/datasets/https-piveau-io-set-data-pre-trained-transformer.jsonld"

ds = Dataset.from_url(url)
ds.download(data_dir="./data")
model, _, _ = ds.load_model(model_dir="./models")
```

```python
from sklearn.metrics import accuracy_score


model.fit(X_train[:2], y_train[:2])

# Predict labels
predictions = model.predict(X_test)


print("Accuracy", accuracy_score(y_test, predictions))
```
